```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
i= 0
urls = [
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=10&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=11&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=12&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=13&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=14&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=15&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=16&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=17&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=18&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=19&sort=PRONAC:asc',
    'http://versalic.cultura.gov.br/#/projetos?limit=12&offset=0&ano_projeto=20&sort=PRONAC:asc',
]
```

```python
p = []
pageCount = 0
pageToScrape = 10
j = 0
```

```python
option = Options()
driver = webdriver.Chrome(options = option)

driver.get(urls[i])

time.sleep(5)

driver.find_element_by_xpath('//*[@id="app-outlet-container"]/app-busca/div/div[3]/div/div[2]/div').click()
driver.find_element_by_xpath('//*[@id="app-outlet-container"]/app-busca/div/div[3]/div/div[2]/div/ul/li[4]/a').click()
```

```python
pageCount = 0
toinput = []
while(pageCount< pageToScrape and i <= len(urls)):
    time.sleep(5)
    projetos = driver.find_elements_by_class_name('card')
    for projeto in projetos:
        area = projeto.find_element_by_xpath('.//div[@class = "card-resposta texto-elipsado"]/div[1]/div[3]/p[2]').text
        data = projeto.find_element_by_xpath('.//div[@class = "card-resposta texto-elipsado"]/div[3]/div[1]/p[2]').text
        municipio = projeto.find_element_by_xpath('.//div[@class = "card-resposta texto-elipsado"]/div[3]/div[2]/p[2]').text
        valor_proposta = projeto.find_element_by_xpath('.//div[@class = "card-resposta texto-elipsado"]/div[4]/div[1]/p[2]').text
        valor_captado = projeto.find_element_by_xpath('.//div[@class = "card-resposta texto-elipsado"]/div[4]/div[2]/p[2]').text
        item = {
            'area': area,
            'data': data,
            'ano': 2010+i,
            'municipio': municipio,
            'valor_proposta': valor_proposta,
            'valor_captado': valor_captado
        }
        toinput.append(item)
    
    pageCount+=1
    element = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH,'//*[@id="app-outlet-container"]/app-busca/div/div[3]/div/div[3]/pagination/ul/li[last()]/a')))
    element.click()
    print('Paginas scrapeadas: ', j)
    j+=1
    time.sleep(5)
    

i+=1
driver.quit()
p+=toinput
        
```

```python
df2 = pd.read_csv('rouanet1020.csv')
df2
```

```python
uflist = []

for i in range(len(df2)):
    splited_string = df2.loc[i, "municipio"].split(', ')
    df2.loc[i, "municipio"] = splited_string[0]
    uflist.append(splited_string[1])


df2["uf"] = uflist
```

```python
df2
```

```python
for i in range(len(df2)):
    df2.loc[i, "valor_proposta"] = df2.loc[i, "valor_proposta"].replace('R$ ', "")
    df2.loc[i, "valor_proposta"] = df2.loc[i, "valor_proposta"].replace('.', "")
    df2.loc[i, "valor_proposta"] = df2.loc[i, "valor_proposta"].replace(',', ".")
    df2.loc[i, "valor_captado"] = df2.loc[i, "valor_captado"].replace('R$ ', "")
    df2.loc[i, "valor_captado"] = df2.loc[i, "valor_captado"].replace('.', "")
    df2.loc[i, "valor_captado"] = df2.loc[i, "valor_captado"].replace(',', ".")


df2
```

```python
df2["valor_captado"] = df2['valor_captado'].astype('float')
df2["valor_proposta"] = df2['valor_proposta'].astype('float')
```
