```python
capitais = ["Fortaleza", "Rio Branco", "Maceió", "Macapá", "Manaus", "Salvador", "Brasília", "Vitória", "Goiania", "São Luís",
            "Cuiabá", "Campo Grande", "Belo Horizonte", "Belém", "João Pessoa", "Curitiba", "Recife", "Teresina", "Rio de Janeiro",
           "Natal", "Porto Alegre", "Boa Vista", "Florianópolis", "São Paulo", "Aracaju", "Palmas"]

capital = []
for index, row in df.iterrows():
    cap = row['municipio']
    if cap in capitais:
        capital.append(1)
    else:
        capital.append(0)
        
df["capital"] = capital
```

```python
class LogisticRegression:
    def __init__(self, lr=0.01, num_iter=100000, fit_intercept=True):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
    
    def __add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def __sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    def __loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        # weights initialization
        self.weight = np.zeros(X.shape[1])
        
        for i in range(self.num_iter):
            z = np.dot(X, self.weight)
            h = self.__sigmoid(z)
            gradient = np.dot(X.T, (h - y)) / y.size
            self.weight -= self.lr * gradient
    
    def predict_prob(self, X):
        if self.fit_intercept:
            X = self.__add_intercept(X)
    
        return self.__sigmoid(np.dot(X, self.weight))
    
    def predict(self, X, threshold=0.5):
        return self.predict_prob(X) >= threshold
```

```python
df.drop(["ano", "data", "municipio"], axis=1, inplace=True)
```

```python
df['area'] = pd.Categorical(df.area)
df['uf'] = pd.Categorical(df.uf)
```

```python
df['areacd'] = df['area'].cat.codes
df['ufcd'] = df['uf'].cat.codes
df["valor_proposta"] = df["valor_proposta"].apply(lambda x: x/1000000)
```

```python
X= df[["areacd", "ufcd"]]
y = df['results']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = LogisticRegression(lr=0.1,num_iter= 100000)
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
preds
```

```python
print("acurácia: ")
print(accuracy_score(y_test,preds))
```
